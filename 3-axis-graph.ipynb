```python
# IMPORTANT LIBRARIES
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.patches import Patch
#import matplotlib.patches as Patch
```

# FOLDER CONNECTION

```python
#TEMPERATURE DATA
#"harorat.csv"
temp = pd.read_csv(r"temp.csv")
temp.head() #just to check either it is read 
```

```python
#WATER RUNOFF
#"C:\Users\Asus\Desktop\Programming\R\lesson_1_python\data\harorat.csv"
runoff = pd.read_csv(r"water_discharge.csv")
runoff.head() #just to check either it is read 
```

```python
#WATER LEVEL
w_level = pd.read_csv(r"water_level.csv")
w_level.head() #just to check either it is read 
```

```python
time = temp['Day']
time.head()
```

```python
#TEMPERATURE GRAPH
fig, ax1 = plt.subplots()
color = 'tab:green'
ax1.plot(time, temp['Tem'],
         linestyle = 'solid', marker = 'o',
         markerfacecolor = 'red', markersize = 3, color=color)
ax1.tick_params(axis='y', labelcolor='darkorange')

plt.xlabel("Days")
plt.ylabel("Temprature")
#background
plt.grid(True, which='both', linestyle='-', linewidth=0.7)
plt.savefig('result1.png', dpi=600, bbox_inches='tight', facecolor='white') 
#plt.show()
```

```python
#Runoff
fig, ax2 = plt.subplots()
color = 'tab:red'
ax2.plot(time, runoff['discharge'], 
         linestyle = 'solid', marker = 'o',
         markerfacecolor = 'black', markersize = 0, color=color)
ax2.tick_params(axis='y', labelcolor='darkorange')

plt.xlabel("Days")
plt.ylabel("Runoff")

#background
plt.grid(True, which='both', linestyle='-', linewidth=0.7)
plt.savefig('result2.png', dpi=600, bbox_inches='tight', facecolor='white') 
#plt.show()
```

```python
#W_level
fig, ax3 = plt.subplots()
color = 'tab:orange'
ax3.plot(time,w_level['w_level'], 
         linestyle = 'solid', marker = 'o',
         markerfacecolor = 'black', markersize = 3, color=color)
ax3.tick_params(axis='y', labelcolor='darkorange')

plt.xlabel("Days")
plt.ylabel("water level")
#background
plt.grid(True, which='both', linestyle='-', linewidth=0.7)
plt.savefig('result3.png', dpi=600, bbox_inches='tight', facecolor='white') 
#plt.show()
```

```python
plt.close('all')  # Close all existing plots
```

# ONE GRAPH FOR 3 PARAMETRES 

```python
fig, ax1 = plt.subplots()
plt.grid(True, which='both', linestyle='-', linewidth=0.5) #background


#COLORS OF VARIABLES 
col_tem = 'red'
col_runoff = 'green'
col_w_level = 'blue'

# First y-axis
ax1.set_xlabel('Days')
ax1.set_ylabel('Temperature °C', color=col_tem)
ax1.plot(time, temp['Tem'], color=col_tem)
ax1.tick_params(axis='y', labelcolor=col_tem)


# Second y-axis
ax2 = ax1.twinx()
ax2.set_ylabel('Runoff m^3/sec', color=col_runoff)
ax2.plot(time, runoff['discharge'], color=col_runoff)
ax2.tick_params( axis='y', labelcolor=col_runoff)

# Third y-axis
ax3 = ax1.twinx()
color = 'tab:green'
ax3.spines['right'].set_position(('outward', 60))
ax3.set_ylabel('Water level sm', color=col_w_level)
ax3.plot(time, w_level['w_level'], color=col_w_level)
ax3.tick_params( axis='y',labelcolor=col_w_level)

# Creating custom legend patches as a list format 
legend_elements = []
legend_elements = [
    Patch(facecolor=col_tem,    edgecolor=col_tem,    label='Temperature'),
    Patch(facecolor=col_runoff, edgecolor=col_runoff, label='Runoff'),
    Patch(facecolor=col_w_level,edgecolor=col_w_level,label='Water level')]

# Adding legend with custom patches to left 
plt.legend(handles=legend_elements, loc='upper left')
plt.xlim([0, 365])
plt.xticks(range(0, 365, 30))

plt.savefig('result_3axis.png', dpi=1000, bbox_inches='tight', facecolor='white') 
plt.show()
```
